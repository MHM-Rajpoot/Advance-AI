## Distil-GPT-2

DistilGPT2 is a smaller, more efficient version of the GPT2 language model. It's a type of artificial intelligence designed for natural language processing tasks. DistilGPT2 uses fewer parameters than GPT2, making it faster and more suitable for devices with limited computing resources. Despite its smaller size, DistilGPT2 still offers impressive language understanding and generation capabilities. It's often used for tasks like text classification, sentiment analysis, and text generation.

```python
from transformers import pipeline
```

```python
generator = pipeline("text-generation", model="distilgpt2")
```

#### 1. Clarity and Specificity

```python
def clear_specific_prompt():
    prompt = "Write a summary of AI history in bullet points."
    result = generator(prompt, max_length=200, num_return_sequences=1, truncation=True)
    print("\n"*2)
    print("Clarity and Specificity Output:")
    print("\n")
    print(result[0]['generated_text'])
    print("\n")
```

```python
clear_specific_prompt()
```

### 2. Context Provision

```python
def context_based_prompt():
    prompt = "As a beginner in AI, explain neural networks in simple terms"
    result = generator(prompt, max_length=250, num_return_sequences=1, truncation=True)
    print("\n"*2)
    print("Context-Based Output:")
    print("\n")
    print(result[0]['generated_text'])
    print("\n")
```

```python
context_based_prompt()
```


### 3. Role-Based Prompting

```python
def role_based_prompt():
    prompt = "Act as a nutritionist and recommend a daily vegetarian meal plan"
    result = generator(prompt, max_length=200, num_return_sequences=1, truncation=True)
    print("\n"*2)
    print("Role-Based Output:")
    print("\n")
    print(result[0]['generated_text'])
    print("\n")
```

```python
role_based_prompt()
```

### 4. Iterative Refinement

```python
def iterative_refinement_prompt():

    # Initial vague prompt
    vague_prompt = "Write about AI"
    vague_result = generator(vague_prompt, max_length=150, num_return_sequences=1, truncation=True)
    print("\n"*2)
    print("Vague Prompt Output:")
    print("\n")
    print(vague_result[0]['generated_text'])
    print("\n")

    # Refined prompt
    refined_prompt = "Write a concise summary of AI applications in healthcare."
    refined_result = generator(refined_prompt, max_length=200, num_return_sequences=1, truncation=True)
    print("\n"*2)
    print("Refined Prompt Output:")
    print("\n")
    print(refined_result[0]['generated_text'])
    print("\n")
```

```python
iterative_refinement_prompt()
```

### 5. Constraints and Boundaries

```python
def constrained_prompt():
    prompt = "Write a 50-word product description for a smartwatch in a professional tone."
    result = generator(prompt, max_length=100, num_return_sequences=1, truncation=True)
    print("\n"*2)
    print("Constrained Prompt Output:")
    print("\n")
    print(result[0]['generated_text'])
    print("\n")
```

**Enhanced conversational abilities:** Claude is designed for more natural and engaging conversations, understanding context and nuance better.

**Increased knowledge:** Claude's training data is more extensive and up-to-date, allowing it to provide more accurate and informative responses.

**Better handling of complex topics:** Claude can tackle more complex and abstract topics, providing thoughtful and well-reasoned responses.

**More transparent and explainable:** Claude is designed to provide clear explanations and justifications for its responses, making it more trustworthy.

Overall, Claude's focus on safety, conversation, and knowledge make it a more advanced and reliable choice compared to DistilGPT2.

```python
import os
import torch
from ctransformers import AutoModelForCausalLM
from huggingface_hub import hf_hub_download

# Check GPU availability and CUDA setup
print("Checking GPU availability...")
if not torch.cuda.is_available():
    print("ERROR: GPU not available. Please set runtime to T4 GPU: Runtime > Change runtime type > T4 GPU")
    exit(1)
else:
    print("GPU detected:", torch.cuda.get_device_name(0))
    print("CUDA version:", torch.version.cuda)

# Download the Claude-inspired model from Hugging Face
model_repo = "TheBloke/Mistral-7B-Claude-Chat-GGUF"
model_file = "mistral-7b-claude-chat.Q4_K_M.gguf"

try:
    print("\nDownloading model...")
    model_path = hf_hub_download(repo_id=model_repo, filename=model_file, local_dir=".")
    print("Model downloaded successfully!")
except Exception as e:
    print(f"Error downloading model: {e}")
    exit(1)

# Load the model with GPU support
try:
    print("\nLoading model with GPU support...")
    llm = AutoModelForCausalLM.from_pretrained(
        model_path,
        model_type="mistral",
        gpu_layers=30  # Reduced to 30 layers to avoid VRAM issues on T4 GPU
    )
    print("Model loaded successfully with GPU support!")
except Exception as e:
    print(f"Error loading model: {e}")
    exit(1)
```

Single effective coding example demonstrating all prompt engineering techniques

```python
# Define the Vicuna 1.1 prompt format used by the model
def format_prompt(prompt):
    return f"A chat between a curious user and an artificial intelligence assistant. The assistant gives helpful, detailed, and polite answers to the user's questions. USER: {prompt} ASSISTANT:"

# Single effective coding example demonstrating all prompt engineering techniques
def prompt_engineering_example():
    # Step 1: Vague prompt (to show iterative refinement)
    vague_prompt = "Write some Python code."
    formatted_vague = format_prompt(vague_prompt)
    try:
        vague_result = llm(formatted_vague, max_new_tokens=150, temperature=0.7)
        print("\n"*2)
        print("\nVague Prompt Output (Lacking Clarity, Context, Role, Constraints):")
        print("\n")
        print(vague_result[:500])  # Truncate for readability
        print("\n")
    except Exception as e:
        print(f"Error in vague prompt: {e}")

    # Step 2: Refined prompt incorporating all techniques
    refined_prompt = (
        "Act as a Python tutor and write a Python function to calculate the nth Fibonacci number, "
        "targeting beginner programmers. Include a concise 50-word explanation, use recursive implementation, "
        "and format the code with clear comments. Ensure the function takes a positive integer n and returns the result."
    )
    formatted_refined = format_prompt(refined_prompt)
    try:
        refined_result = llm(formatted_refined, max_new_tokens=300, temperature=0.7)
        print("\n"*2)
        print("Refined Prompt Output (Clarity, Context, Role-Based, Constraints):")
        print("\n")
        print(refined_result[:500])  # Truncate for readability
        print("\n")
    except Exception as e:
        print(f"Error in refined prompt: {e}")

    # Check GPU utilization after inference
    print("GPU utilization after inference:")
    !nvidia-smi

def main():
    print("\n=== Prompt Engineering Demo with Claude-Inspired Model on GPU ===\n")
    print("Single Coding Example: Generating a Fibonacci Function in Python\n")
    prompt_engineering_example()

# Execute the main function
if __name__ == "__main__":
    main()
```
